```python
import pandas as pd
```

```python
df = pd.read_csv('df_news_translators.csv')
df
```

```python
df = df.filter(['burmese', 'english'], axis=1)
df
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df[df.english.str.contains('//')]
```

```python
def augment_df(df):
    burmese_list = []
    english_list = []
    for i in range(len(df)):
        if len(df.burmese.iloc[i].split("//")) == len(df.english.iloc[i].split("//")):
            for j in range(len((df.burmese.iloc[i].split("//")))):
                burmese_list.append(df.burmese.iloc[i].split("//")[j])
                english_list.append(df.english.iloc[i].split("//")[j])
    data_frame = {"burmese": burmese_list, "english": english_list}
    df_align = pd.DataFrame(data_frame)
    return df_align
```

```python
df_align = augment_df(df)
df_align
```
